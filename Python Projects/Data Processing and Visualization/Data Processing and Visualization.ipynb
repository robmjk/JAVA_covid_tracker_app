```python
""" 
Csv data analysis

Reads the species_taxon.csv file and will identify all the categories used and count how many times each appears, then 
creating a pie chart and outputting this information into Taxon-Jan2018.csv.
"""

import csv

str_category = "Taxon"
file_infile = open("species_taxon.csv", "r",encoding='utf-8')
csv_dict_reader = csv.DictReader(file_infile)

# Create lists for storing the unique values & count of each
lst_categories = []
lst_counts = []
total_count = 0

# Iterate through the file (which is open in the reader) by
# row which is a list of column values. 
for lst_cols in csv_dict_reader:
     # if lst_cols[str_category] isnt in unique list, 
     # append to cols[category] value to unique list
    if lst_cols[str_category] not in lst_categories:
        # Add that weve seen on to our unique categories
        #     and add the number of times weve seen to the 
        #     counts list.
        lst_categories.append(lst_cols[str_category])
        lst_counts.append(1)
    else:
        # We have it in our list of categories, record we 
        #   have seen another occurrence.
        idx = lst_categories.index(lst_cols[str_category])
        lst_counts[idx] = lst_counts[idx] + 1
        total_count = total_count + 1

file_infile.close()
print("Total Count: " + str(total_count))
      
# open the file for writing as a csv
cat_file = open('Taxon-Jan2018.csv', mode='w') 
cat_writer = csv.writer(cat_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

# Print out the unique categories and count per category
for idx in range(0,len(lst_categories)):
    print(str(lst_categories[idx]) + ": " + str(lst_counts[idx]))
    # 'value', absolute number, percentage
    cat_percent = round(lst_counts[idx] / total_count * 100,2)
    print(cat_percent)
    cat_writer.writerow([str(lst_categories[idx]), str(lst_counts[idx]), str(cat_percent)+"%"])
    
import matplotlib.pyplot as plt
    
# Data to plot
labels = lst_categories
sizes = lst_counts

# Plot
# plt.pie(sizes, labels=labels, autopct='%1.1f%%')
patches, texts = plt.pie(sizes)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
#plt.tight_layout()
plt.show()
```
